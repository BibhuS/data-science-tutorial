```python
import pandas as pd
import numpy as np
```

```python
#let's create a dataframe
df1 = pd.DataFrame(np.random.randn(5,3),columns=['a','b','c'])
df1
df2 = pd.DataFrame(np.random.randn(2,3),index=[6,7],columns=['d','e','f'])
df2
df3 = pd.concat([df1,df2],axis=1)
df3
```

```python
df3['a'].isna()
```

```python
pd.isna(df3['d'])
```

```python
# inserting some missing data
a = df1.copy()
a
a.loc[2:3,'c'] = np.nan
a
a.loc[4,'c'] = None
a
```
