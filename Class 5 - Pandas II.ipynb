```python
df1 = pd.DataFrame({'a':[1,2,3,4],
                    'b':[5,6,7,8],
                    'c':[9,10,11,12]})
#df2 = pd.DataFrame({'a':[5,6,1,1],
#                    'b':[5,7,2,0],
#                    'c':[9,4,2,7]})
#df3 = pd.concat([df1,df2],keys=['first','second'])
df1
#df3.loc['second']
#repeating the keys. how?
```

```python
sr1 = pd.Series([20,21,22,23])
sr1
```

```python
df2 = pd.concat([df1,sr1],axis=1)
df2
```

```python
sr2 = pd.Series([41,42,43,44],name='d')
df2 = pd.concat([df1,sr2],axis=1)
df2
```

```python
df4 = pd.DataFrame({'a':[5,6,4],'b':[4,1,9]},index=[2,3,5])
df4
```

```python
#pd.concat([df1,df4])
pd.concat([df1,df4],axis=1)
#pd.concat([df1,df4],ignore_index=True)
```

```python
df6 = pd.concat([df1,df4],axis=1,join='inner')
df6
```

```python
pd.merge(df1,df4,how='inner',left_index=True,right_index=True)
```

```python
pd.merge(df1,df4,how='left',left_index=True,right_index=True)
```

```python
pd.merge(df1,df4,how='right',left_index=True,right_index=True)
```

```python
pd.merge(df4,df1,how='outer',left_index=True,right_index=True)
```
