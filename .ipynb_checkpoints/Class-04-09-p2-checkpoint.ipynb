```python
def create_dicn(train_dir) :
    emails = [os.path.join(train_dir,f) for f in os.listdir(train_dir)]
    all_words = []
    for mail in emails :
        with open(mail) as m:
            for i,line in enumerate(m):
                if i == 2:
                    words = line.split()
                    all_words+=words
    dictionary = cl.Counter(all_words)
    print(dictionary)
    data_to_remove = list(dictionary.keys())
    for i in range(0,len(data_to_remove)):
        if data_to_remove[i].isalpha() == False:
            del dictionary[data_to_remove[i]]
        elif len(data_to_remove[i]) == 1:
            del dictionary[data_to_remove[i]]
    dictionary = dictionary.most_common(3000)
    return dictionary  

#create_dicn('data/ling-spam/train-mails')
```
