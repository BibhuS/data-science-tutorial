```python
# Required Python Machine learning Packages
import pandas as pd
import numpy as np
# For preprocessing the data
from sklearn.preprocessing import Imputer
from sklearn import preprocessing
# To split the dataset into train and test datasets
from sklearn.cross_validation import train_test_split
# To model the Gaussian Navie Bayes classifier
from sklearn.naive_bayes import GaussianNB
# To calculate the accuracy score of the model
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data/census/census.csv',header = None)
df.head()
```

```python
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                    'marital_status', 'occupation', 'relationship',
                    'race', 'sex', 'capital_gain', 'capital_loss',
                    'hours_per_week', 'native_country', 'income']
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.age == 40 & df.education == 'Assoc-voc']
```

```python
for value in ['workclass', 'education',
          'marital_status', 'occupation',
          'relationship','race', 'sex',
          'native_country', 'income']:
    print(value,":", sum(df[value] == '?'))
```

```python
# Taking a back up of dataframe
df_bkup = df
```

```python
df_bkup.describe(include= 'all')
```

```python
for value in ['workclass', 'education',
          'marital_status', 'occupation',
          'relationship','race', 'sex',
          'native_country', 'income']:
    df_bkup[value].replace(['?'], [df_bkup.describe(include='all')[value][2]],inplace=True)
```

```python
df_bkup.info()
```

```python
le = preprocessing.LabelEncoder()
workclass_cat = le.fit_transform(df.workclass)
education_cat = le.fit_transform(df.education)
marital_cat   = le.fit_transform(df.marital_status)
occupation_cat = le.fit_transform(df.occupation)
relationship_cat = le.fit_transform(df.relationship)
race_cat = le.fit_transform(df.race)
sex_cat = le.fit_transform(df.sex)
native_country_cat = le.fit_transform(df.native_country)
```

```python
#initialize the encoded categorical columns
df_bkup['workclass_cat'] = workclass_cat
df_bkup['education_cat'] = education_cat
df_bkup['marital_cat'] = marital_cat
df_bkup['occupation_cat'] = occupation_cat
df_bkup['relationship_cat'] = relationship_cat
df_bkup['race_cat'] = race_cat
df_bkup['sex_cat'] = sex_cat
df_bkup['native_country_cat'] = native_country_cat
```

```python
#drop the old categorical columns from dataframe
dummy_fields = ['workclass', 'education', 'marital_status', 
                  'occupation', 'relationship', 'race',
                  'sex', 'native_country']
df_bkup = df_bkup.drop(dummy_fields, axis = 1)
```

```python
df_bkup = df_bkup.reindex(['age', 'workclass_cat', 'fnlwgt', 'education_cat',
                                    'education_num', 'marital_cat', 'occupation_cat',
                                    'relationship_cat', 'race_cat', 'sex_cat', 'capital_gain',
                                    'capital_loss', 'hours_per_week', 'native_country_cat', 
                                    'income'], axis= 1)

df_bkup.head()
```

```python
# normalize the column vectors
num_features = ['age', 'workclass_cat', 'fnlwgt', 'education_cat', 'education_num',
                'marital_cat', 'occupation_cat', 'relationship_cat', 'race_cat',
                'sex_cat', 'capital_gain', 'capital_loss', 'hours_per_week',
                'native_country_cat']

scaled_features = {}
for each in num_features:
    mean, std = df_bkup[each].mean(), df_bkup[each].std()
    scaled_features[each] = [mean, std]
    df_bkup.loc[:, each] = (df_bkup[each] - mean)/std
```

```python
features = df_bkup.values[:,:14]
target = df_bkup.values[:,14]
features_train, features_test, target_train, target_test = train_test_split(features,
                                                                            target, test_size = 0.33, random_state = 10)
```

```python
model = GaussianNB()
model.fit(features_train, target_train)
target_pred = model.predict([features_test])
target_pred
```

```python
accuracy_score(target_test, target_pred, normalize = True)
```
