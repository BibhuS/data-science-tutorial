```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read dataset
df = pd.read_csv('data/Pokemon.csv', index_col=0)
df.head()
```

```python
df['Type 1'].unique()
```

```python
#visualizing using seaborn
sns.lmplot(data=df,x='Attack',y='Defense')
```

```python
#what if we don't need to show the regression line?
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False)
```

```python
#let's bring another piece of information where we indicate the stage by showing dots with different colors
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False,hue='Stage')
```

```python
#let's build a box and whiskers plot
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

```python
#let's build a box and whiskers plot
df1 = df.drop(['Total','Stage','Legendary'],axis=1)
sns.boxplot(data=df1)
plt.xticks(rotation=90)
```

```python
# Set theme
sns.set_style('darkgrid')

#other color themes
# whitegrid
# dark
# white
# tile
```

```python
#setting your own color palette
fig,ax=plt.subplots(figsize=(15.0,5.0))
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]


# Violin plot with Pokemon color palette
sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors) # Set color palette
```

```python
# Violin plot with Pokemon color palette
fig,ax=plt.subplots(figsize=(15.0,5.0))
sns.swarmplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors) # Set color palette
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(15,6))
 
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.5) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type')
```

```python
# Calculate correlations
corr = df1.corr()
 
# Heatmap
sns.heatmap(corr)
```

```python
# Distribution Plot (a.k.a. Histogram)
sns.distplot(df.Attack)
```

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

```python
# Factor Plot
g = sns.factorplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
```

```python
# Joint Distribution Plot
sns.jointplot(x='Attack', y='Defense', data=df)
```
